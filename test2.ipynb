```python
from PIL import Image
from PIL.Image import Resampling
from PIL import ImageDraw
from PIL import ImageFont
from PIL import ImageEnhance

width = 1280
height = 720
opacity = 1
text = 'DOMINGO DA '
filename = 'result.png'
black = (0,0,0)
white = (255,255,255)
transparent = (0,0,0,0)
fontWidthScale = 0.95
fontHeightScale = 1.1

font = ImageFont.truetype('Montserrat-Bold.ttf', 72)
wm = Image.new('RGBA',(width,height),transparent)
im = Image.new('RGBA',(width,height),transparent) # Change this line too.

draw = ImageDraw.Draw(wm)
w,h = draw.textsize(text, font, spacing=4)
draw.text((0, 0),text,white,font)
wm = wm.resize((int(wm.size[0] * fontWidthScale), int(wm.size[1] * fontHeightScale)), Resampling.BILINEAR)

en = ImageEnhance.Brightness(wm)
#en.putalpha(mask)
mask = en.enhance(1-opacity)
im.paste(wm,(25,25),mask)

im.save(filename)
```
